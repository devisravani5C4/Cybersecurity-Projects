```python
import cv2
import string
import os
from Crypto.Cipher import AES
from Crypto.Util.Padding import pad,unpad
from Crypto.Random import get_random_bytes
import hashlib
```

```python
#key Generation

def derive_key(userKey):
  return hashlib.sha256(userKey.encode()).digest()[:16]
```

```python
#encryption of the text message using key

def encrypt_message(msg,userKey):
  #hashing key
  key=derive_key(userKey)
  cipher=AES.new(key,AES.MODE_CBC)
  ct=cipher.encrypt(pad(msg.encode(),AES.block_size))
  return cipher.iv + ct
```

```python
#decryption of the text message using key

def decrypt_message(cipher_bytes,userKey):
  #hashing key
  key=derive_key(userKey)
  iv=cipher_bytes[:16]
  ct=cipher_bytes[16:]
  cipher=AES.new(key,AES.MODE_CBC,iv)
  return unpad(cipher.decrypt(ct),AES.block_size).decode()
```

These 3 functions helps in encrypting and decrypting the actual message.

```python
x_enc=x.copy()
n=0
m=0
z=0
kl=0
for i in range(l):
  org_val=x_enc[n,m,z]
  new_val=encrypted_bytes[i]^ca[Key[kl]]
  x_enc[n,m,z]=new_val
  print("original value is ",org_val," after performing XOR operation at pixel (",n,m,z,") i.e., ", encrypted_bytes[i]," XOR ",ca[Key[kl]]," the new value is ",new_val)
  n=n+1
  m=m+1
  m=(m+1)%3
  z=(z+1)%3
  kl=(kl+1)%len(Key)
```
