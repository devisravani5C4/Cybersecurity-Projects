```python

import numpy as np
import cv2
```

```python
#characters to ascii values and vice versa
ca={chr(i):i for i in range(255)}
ac={i:chr(i) for i in range(255)}
```

```python
print(ca)
print(ac)
```

Defining the key and message:-

```python
text="hello"
key="123"
```

```python
#random pixel value image

x=np.random.randint(0,255,(10,10,3),dtype=np.uint8)
```

```python
x
```

This is my cover image.

```python
x.shape
```

```python
text_ascii=[ca[ch] for ch in text]
key_ascii=[ca[ch] for ch in key]
```

```python
print(text,"=",text_ascii)

print(key,"=",key_ascii)
```

ENCRYPTION:-(Hiding message into cover image)

```python
x_enc=x.copy()
n,m,z=0,0,0
l=len(text)
kl=0
for i in range(l):
  char_val=ca[text[i]]^ca[key[kl]]
  for bit_pos in range(8):
     bit=(char_val >> (7-bit_pos))&1
     org_val=x_enc[n,m,z]
     x_enc[n,m,z]=(org_val & 0xFE)|bit

     print("original value is ",org_val," after embedding bit ",bit ," at pixel (",n,m,z,") of ",text[i]," the new value is ",x_enc[n,m,z])
     z=(z+1)%3
     if z==0:
       m=m+1
       if m==x_enc.shape[1]:
          m=0
          n=n+1
  kl=(kl+1)%len(key)
```

```python
x_enc
```

```python
cv2.cvtColor(x_enc,cv2.COLOR_BGR2RGB)
```

Now this is my stego-object (cover+text message)

It is ready at sender side

DECRYPTION:-

(at reciever side, the stegno-object is recieved . Now we have to decrypt the image and get the actual text message as shown below.)

```python
n,m,z=0,0,0
kl=0
decrypt_msg=""
for i in range(l):
  value=0
  for bit_pos in range(8):
    bit=x_enc[n,m,z] & 1
    value=(value << 1)| bit
    print("Reading bit ",bit," from pixel (",n,m,z,")")
    z=(z+1)%3
    if z==0:
      m=m+1
      if m==x_enc.shape[1]:
         m=0
         n=n+1

  org_char=ac[value ^ ca[key[kl]]]
  print("decrypted value =",value," XOR ",ca[key[kl]]," =",value^ca[key[kl]]," i e., ",org_char)
  decrypt_msg=decrypt_msg+org_char
  kl=(kl+1)%len(key)

print(decrypt_msg)
```
